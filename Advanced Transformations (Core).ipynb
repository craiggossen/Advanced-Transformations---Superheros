### Clean Superhero Info

```python
info[['Hero','Publisher']] =info['Hero|Publisher'].str.split('|',expand=True)
info.drop(columns='Hero|Publisher', inplace = True)
```

```python
#Isolating one instance
measure = info['Measurements'][0]
print(type(measure))
measure
```

```python
#Replacing to align with json formatting
measure = measure.replace("'",'"')
measure
```

```python
## now we can use json.loads
measure_loads = json.loads(measure)
print(type(measure_loads))
measure_loads
```

```python
#Apply to all data in the column

info['Measurements'] = info['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
info['Measurements'] = info['Measurements'].apply(json.loads)
info['Measurements'].head()
```

```python
info.head()
```

```python
#Split dict into two separate cols
measurements = info['Measurements'].apply(pd.Series)
measurements
```

```python
#Concatenate new cols onto the dataframe
info = pd.concat((info, measurements), axis = 1)
info.head()
```

```python
#Drop original column
info.drop(columns='Measurements', inplace = True)
info.head()
```

```python
#Rename cols
info['Height (cm)'] = info['Height']
info['Weight (kg)'] = info['Weight']
```

```python
info.head()
```

```python
#Removing unnecessary strings in data to get to numeric data
info['Height (cm)'] = info['Height (cm)'].map(lambda x: x.replace(' cm',""))
info['Weight (kg)'] = info['Weight (kg)'].map(lambda x: x.replace(' kg',""))
```

```python
#Drop original cols
info.drop(columns=['Height', 'Weight'], inplace = True)
```

```python
info.head()
```

```python
#Check instance for dtype
weight = info['Weight (kg)'][0]
print(type(weight))
weight
```

```python
#Changing dtype to numeric
info[['Height (cm)', 'Weight (kg)']] = info[['Height (cm)', 'Weight (kg)']].apply(pd.to_numeric)
info.head()
```

```python
#Checking type to ensure it is numeric
weight = info['Weight (kg)'][0]
print(type(weight))
weight
```

```python
shp['Powers New'] = shp['Powers'].str.split(',', expand = False)
```

```python
## exploding the column of Powers/list within
exploded = shp.explode('Powers New')
exploded
```

```python
## saving the unique values from the exploded column
cols_to_make = exploded['Powers New'].dropna().unique()

sorted_cols_to_make = sorted(cols_to_make)
type(sorted_cols_to_make)
```

```python
sorted_cols_to_make
```

```python
for col in sorted_cols_to_make:
    shp[col] = shp['Powers'].str.contains(col)
shp.head()
```

```python
shp.drop(columns=['Powers', 'Powers New'], inplace = True)
```

### Combine the dataframes


```python
#Dataframes are shp and info
```

```python
merged = pd.merge(info, shp, left_on='Hero', right_on='hero_names', how = 'inner')
```

```python
merged.head()
```

### Questions

Compare the average weight of super powers who have Super Speed to those who do not.

```python
#Applying filter for SS heros
ss_weight = merged['Super Speed'] == True
dfss = merged[ss_weight]
ss_amount = dfss['Weight (kg)'].mean()

print(f'The average weight of a hero with Super Speed is {ss_amount:,.0f} kg.')
```

```python
#Applying filter for non-SS heros
noss_weight = merged['Super Speed'] == False
dfnoss = merged[noss_weight]
noss_amount = dfnoss['Weight (kg)'].mean()

print(f'The average weight of a hero without Super Speed is {noss_amount:,.0f} kg.')
```
